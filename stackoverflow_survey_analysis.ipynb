# Stack Overflow Developer Survey Analysis

This notebook analyzes the 2025 Stack Overflow Developer Survey dataset using Python and Pandas.
The analysis focuses on participant count, complete responses, work experience, remote work, Python usage, learning methods, compensation, education level, age groups, and industries.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("results.csv", low_memory=False)
df.head()
```

```python
df.shape
```

```python
df.columns.tolist()
```

## 1. How many people participated in the survey?

Each row in the dataset represents one survey respondent. Therefore, the total number of participants is equal to the number of rows in the dataset.

```python
total_participants = df.shape[0]
total_participants
```

```python
participant_summary = pd.DataFrame({
    "Metric": ["Total Participants"],
    "Value": [df.shape[0]]
})

participant_summary
```

## 2. How many participants answered all questions?

A complete response means that the participant has no missing values across all columns in the dataset.
Rows with no missing values were counted using `dropna()`.

```python
complete_responses = df.dropna().shape[0]
complete_responses
```

```python
complete_response_summary = pd.DataFrame({
    "Metric": ["Complete Responses"],
    "Value": [df.dropna().shape[0]]
})

complete_response_summary
```

## 3. What are the central tendency measures for WorkExp?

The central tendency of participants' work experience was analyzed using mean, median, and mode.
Missing values were ignored during the calculation.

```python
workexp_mean = df["WorkExp"].mean()
workexp_median = df["WorkExp"].median()
workexp_mode = df["WorkExp"].mode()[0]

workexp_mean, workexp_median, workexp_mode
```

```python
workexp_summary = pd.DataFrame({
    "Measure": ["Mean", "Median", "Mode"],
    "Value": [
        df["WorkExp"].mean(),
        df["WorkExp"].median(),
        df["WorkExp"].mode()[0]
    ]
})

workexp_summary
```

## 4. How many people work remotely?

The `RemoteWork` column was used to identify respondents who work remotely.
First, the distribution of all remote work categories was checked, then the number of remote workers was calculated.

```python
df["RemoteWork"].value_counts(dropna=False)
```

```python
remote_workers = df[df["RemoteWork"] == "Remote"].shape[0]
remote_workers
```

```python
remote_work_summary = pd.DataFrame({
    "Metric": ["Remote Workers"],
    "Value": [remote_workers]
})

remote_work_summary
```

## 5. What percentage of participants program with Python?

The `LanguageHaveWorkedWith` column was used to identify respondents who have worked with Python.
Rows containing `Python` in this column were counted and divided by the total number of participants.

```python
python_users = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)

python_percentage = python_users.mean() * 100
python_percentage
```

```python
python_users = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)

python_summary = pd.DataFrame({
    "Metric": ["Python Users", "Python Usage Percentage"],
    "Value": [
        python_users.sum(),
        round(python_users.mean() * 100, 2)
    ]
})

python_summary
```

## 6. How many people learned programming through online courses?

The `LearnCode` column was used to identify respondents who learned programming through online courses or certifications.
Rows containing `Online Courses or Certification` were counted.

```python
online_course_learners = df["LearnCode"].str.contains("Online Courses or Certification", na=False)

online_course_count = online_course_learners.sum()
online_course_count
```

```python
online_course_summary = pd.DataFrame({
    "Metric": ["Online Course Learners"],
    "Value": [online_course_count]
})

online_course_summary
```

## 7. What are the mean and median yearly compensation values by country among participants who program with Python?

Participants who have worked with Python were filtered using the `LanguageHaveWorkedWith` column.
Then, the data was grouped by country and the mean and median values of `ConvertedCompYearly` were calculated for each country.
Missing compensation values were ignored automatically during aggregation.

```python
python_users_df = df[df["LanguageHaveWorkedWith"].str.contains("Python", na=False)]

python_users_df.head()
```

```python
python_comp_by_country = (
    python_users_df
    .groupby("Country")["ConvertedCompYearly"]
    .agg(["mean", "median"])
    .reset_index()
)

python_comp_by_country
```

```python
python_users_df = df[df["LanguageHaveWorkedWith"].str.contains("Python", na=False)]

python_comp_by_country = (
    python_users_df
    .groupby("Country")["ConvertedCompYearly"]
    .agg(
        Respondent_Count="count",
        Mean_Yearly_Compensation="mean",
        Median_Yearly_Compensation="median"
    )
    .reset_index()
)

python_comp_by_country["Mean_Yearly_Compensation"] = python_comp_by_country["Mean_Yearly_Compensation"].round(2)
python_comp_by_country["Median_Yearly_Compensation"] = python_comp_by_country["Median_Yearly_Compensation"].round(2)

python_comp_by_country_sorted = python_comp_by_country.sort_values(
    by="Median_Yearly_Compensation",
    ascending=False
)

python_comp_by_country_sorted
```

```python
python_comp_by_country = (
    python_users_df
    .groupby("Country")["ConvertedCompYearly"]
    .agg(["mean", "median"])
    .reset_index()
)

python_comp_by_country
```

## 8. What are the education levels of the top 5 participants with the highest compensation?

The dataset was sorted by `ConvertedCompYearly` in descending order.
The top 5 respondents with the highest yearly compensation were selected, and their education levels were displayed using the `EdLevel` column.

```python
top_5_compensation = df.sort_values(
    by="ConvertedCompYearly",
    ascending=False
).head(5)

top_5_compensation[["ConvertedCompYearly", "EdLevel"]]
```

```python
top_5_education = (
    df.sort_values(by="ConvertedCompYearly", ascending=False)
    .head(5)[["ConvertedCompYearly", "EdLevel"]]
    .reset_index(drop=True)
)

top_5_education
```

## Bonus 1. What percentage of participants program with Python in each age group?

The `Age` column was used to group participants by age group.
For each age group, the percentage of respondents who have worked with Python was calculated using the `LanguageHaveWorkedWith` column.

```python
df["UsesPython"] = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)

df[["Age", "LanguageHaveWorkedWith", "UsesPython"]].head()
```

```python
python_by_age = (
    df.groupby("Age")["UsesPython"]
    .mean()
    .mul(100)
    .round(2)
    .reset_index()
)

python_by_age.columns = ["Age Group", "Python Usage Percentage"]

python_by_age
```

## Bonus 2. What are the most common industries among remote workers in the top 25% yearly compensation group?

The 75th percentile of `ConvertedCompYearly` was calculated to identify the top 25% compensation group.
Then, respondents who were both in this top compensation group and working remotely were filtered.
Finally, the most common industries were identified using the `Industry` column.

```python
compensation_75th_percentile = df["ConvertedCompYearly"].quantile(0.75)
compensation_75th_percentile
```

```python
top_25_remote_workers = df[
    (df["ConvertedCompYearly"] >= compensation_75th_percentile) &
    (df["RemoteWork"] == "Remote")
]

top_25_remote_workers.head()
```

```python
top_industries_table = (
    top_25_remote_workers["Industry"]
    .value_counts()
    .head(10)
    .reset_index()
)

top_industries_table.columns = ["Industry", "Count"]

top_industries_table
```
